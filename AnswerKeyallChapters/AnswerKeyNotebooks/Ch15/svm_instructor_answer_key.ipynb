# Support Vector Machine (SVM) Lab  
## Instructor Answer Key Notebook

**Objective:** Build a classification model using Support Vector Machine (SVM) to predict customer purchase behavior.

This notebook includes:
- dataset creation
- preprocessing
- train-test split
- feature scaling
- SVM model training
- evaluation
- simple visualization

## Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Step 2: Create the Dataset

```python
data = {
    "Age": [25, 35, 45, 20, 30, 22, 40, 28, 50, 23],
    "Income": [30, 60, 65, 20, 45, 25, 70, 48, 80, 27],
    "Purchase": [0, 1, 1, 0, 1, 0, 1, 1, 1, 0]
}

df = pd.DataFrame(data)
df
```

## Step 3: Define Features and Target

```python
X = df[["Age", "Income"]]
y = df["Purchase"]

print("Features:")
print(X.head())
print("\nTarget:")
print(y.head())
```

## Step 4: Split the Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))
```

## Step 5: Feature Scaling

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("Scaled training data:")
print(X_train[:5])
```

## Step 6: Train the SVM Model

```python
model = SVC(kernel="linear")
model.fit(X_train, y_train)
```

## Step 7: Make Predictions

```python
y_pred = model.predict(X_test)
print("Predictions:", y_pred)
print("Actual:", y_test.values)
```

## Step 8: Evaluate the Model

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))
```

## Step 9: Visualize the Dataset

```python
plt.figure(figsize=(7, 5))
plt.scatter(df["Age"], df["Income"], c=df["Purchase"])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Customer Purchase Dataset")
plt.show()
```

## Step 10: Try Another Kernel

```python
rbf_model = SVC(kernel="rbf", gamma="scale")
rbf_model.fit(X_train, y_train)
rbf_pred = rbf_model.predict(X_test)

print("RBF Kernel Accuracy:", accuracy_score(y_test, rbf_pred))
print("\nRBF Confusion Matrix:\n", confusion_matrix(y_test, rbf_pred))
```

## Instructor Notes

- SVM performs well when features are scaled properly.
- The linear kernel is useful when the classes are approximately linearly separable.
- The RBF kernel can capture nonlinear boundaries.
- Students can be asked to compare kernels and explain the effect of scaling.
