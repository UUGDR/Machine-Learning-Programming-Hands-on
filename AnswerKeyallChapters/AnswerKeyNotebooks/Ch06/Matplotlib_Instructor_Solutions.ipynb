```
import pandas as pd
import matplotlib.pyplot as plt
```

```
from io import StringIO

data = StringIO("""speed,dist
4,2
4,10
7,4
7,22
8,16
9,10
10,18
10,26
11,17
11,28
12,14
12,20
13,26
13,34
14,46
15,26
15,36
16,32
16,40
17,32
17,40
18,42
18,56
19,76
20,48
20,52
22,84
23,92
24,93
24,120
""")

cars = pd.read_csv(data)
cars.head()
```

```
bins = [0, 10, 15, 20, 100]
labels = ["0–10", "11–15", "16–20", "21+"]

cars["speed_group"] = pd.cut(cars["speed"], bins=bins, labels=labels)
```

```
group_avg = cars.groupby("speed_group")["dist"].mean()

plt.bar(group_avg.index, group_avg.values)
plt.title("Average Stopping Distance by Speed Group")
plt.show()
```

```
plt.hist(cars["speed"], bins=6)
plt.show()
```

```
group_counts = cars["speed_group"].value_counts().sort_index()

plt.pie(group_counts.values, labels=group_counts.index, autopct="%1.1f%%")
plt.show()
```
