# K-Means Clustering Lab (Instructor Answer Key)

## Problem Statement

An online retail company wants to group its customers into different segments based on their annual spending score and purchase frequency. The marketing team plans to use these groups for targeted promotions, personalized recommendations, and customer retention strategies.

Since the customer groups are not known in advance, **K-Means clustering** can be used to discover natural patterns in the data.

In this lab, we build a K-Means clustering model using **Scikit-Learn**, visualize the clusters, identify cluster centroids, and use the **Elbow Method** to estimate the best number of clusters.

After completing this lab, students should be able to:

1. understand the purpose of clustering in machine learning  
2. generate and visualize a dataset for clustering

4. interpret cluster labels and cluster centroids  
5. use the Elbow Method to estimate an appropriate value of **K**  
6. answer conceptual and practical questions related to clustering


## Step 1: Import Required Libraries

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

## Step 2: Generate a Customer-like Dataset

```
X, y_true = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=0.75,
    random_state=42
)

print("Shape of dataset:", X.shape)
print(X[:5])
```

## Step 3: Visualize the Dataset

```
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Customer Dataset for Clustering", fontsize=14)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Step 4: Create and Train the K-Means Model

```
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

kmeans.fit(X)

cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("First 10 cluster labels:")
print(cluster_labels[:10])

print("\nCluster centroids:")
print(centroids)
```

## Step 5: Plot the Clustered Data and Centroids

```
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, s=50, cmap="viridis")
plt.scatter(centroids[:, 0], centroids[:, 1], s=250, marker="X", c="red", label="Centroids")
plt.title("K-Means Clustering Result", fontsize=14)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

## Step 6: Predict the Cluster for New Customers

```
new_points = np.array([[0, 5], [-6, -7], [5, 2]])

predicted_clusters = kmeans.predict(new_points)

print("New Data Points:")
print(new_points)

print("\nPredicted Cluster Labels:")
print(predicted_clusters)
```

## Step 7: Apply the Elbow Method

```
wcss = []

for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X)
    wcss.append(model.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker="o")
plt.title("Elbow Method", fontsize=14)
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.xticks(range(1, 11))
plt.show()
```

## Step 8: Interpretation

**Sample Answers**

1. The elbow generally appears around **K = 3**.  
2. WCSS decreases as **K** increases because points are assigned to smaller and more specific clusters, reducing the distance to their centroids.  
3. A very large value of **K** may over-segment the data, reduce interpretability, and produce clusters that are not meaningful in practice.


## Concept Questions

**Answer Key**

1. **K-Means is an unsupervised learning algorithm** because it groups data without using predefined class labels.  
2. A **centroid** is the center of a cluster. It represents the mean location of all points assigned to that cluster.  
3. **WCSS** measures the total squared distance between data points and their assigned cluster centroids.  
4. One limitation is that the elbow point may not always be clearly visible, making the choice of **K** subjective.  
5. Applications include **customer segmentation, document grouping, image compression, anomaly detection, and market analysis**.
